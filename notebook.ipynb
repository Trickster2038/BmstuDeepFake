```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("dataframe.csv")
df
```

```python
df
```

```python
raw = df.filter(like='_raw', axis=1)

for col in raw:
    name = str(col).split('_')[0] + str(col).split('_')[1] + '_' + 'std'
    df[name] = df[str(col)].map(lambda x: 
        np.array(list(
            map(lambda y: int(y), 
                 list(filter(lambda z: z != '', 
                             x.strip().replace('[', '').replace(']', '').split(' ')))
        )) 
    ).std())

df
```
