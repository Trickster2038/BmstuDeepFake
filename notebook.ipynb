```python
import pandas as pd
import numpy as np
# import seaborn as sns
```

```python
# ? check empty faces ?

df = pd.read_json("../dataframe_total.json")
# df = df.sample(50)
df = df[df['overall_face_raw'].map(lambda x: len(x)) > 0]
df.head(5)
```

```python
# %%capture --no-stderr --output

raw = df.filter(like='_raw', axis=1).copy()

print('\nstage 0')

for col in raw:
    print('+', end='')
    name = str(col).split('_')[0] + '_' + str(col).split('_')[1] + '_std'
    df[name] = raw[str(col)].map(lambda x: np.array(x).std())

print('\nstage 1')
    
raw = df.filter(like='_raw', axis=1).copy()

for col in raw:
    print('+', end='')
    name = str(col).split('_')[0] + '_' + str(col).split('_')[1] + '_var'
    df[name] = raw[str(col)].map(lambda x: np.array(x).var())

print('\nstage 2')

def hist_split(arr):
    res = [0] * 256
    for x in arr:
        res[x] += 1
    return res

raw = df.filter(like='_raw', axis=1).copy()

for col in raw:
    print('+', end='')
    name = str(col).split('_')[0] + '_' + str(col).split('_')[1] + '_hist_simple'
    df[name] = raw[str(col)].map(lambda x: hist_split(np.array(x)))

print('\nstage 3')

hist= df.filter(like='_hist_simple', axis=1).copy()

for col in hist:
    print('+', end='')
    name = str(col).split('_')[0] + '_' + str(col).split('_')[1] + '_hist_std'
    df[name] = hist[str(col)].map(lambda x: np.array(x).std())

print('\nstage 4')

# hist= df.filter(like='_hist', axis=1).copy()

for col in hist:
    print('+', end='')
    name = str(col).split('_')[0] + '_' + str(col).split('_')[1] + '_hist_var'
    df[name] = hist[str(col)].map(lambda x: np.array(x).var())


print('\nstage 5')

def noise_percent(arr, percent_threshold=1, ignore_zero=True):
    threshold = sum(arr) * percent_threshold / 100
    cnt = 0
    for x in arr:
        if (x < threshold and ((x != 0) or (not ignore_zero))):
            cnt += 1
    return cnt * 100 / float(sum(arr))

for col in hist:
    print('+', end='')
    name = str(col).split('_')[0] + '_' + str(col).split('_')[1] + '_hist_noise_0.5'
    df[name] = hist[str(col)].map(lambda x: noise_percent(np.array(x), 0.5))

print('\nstage 6')

for col in hist:
    print('+', end='')
    name = str(col).split('_')[0] + '_' + str(col).split('_')[1] + '_hist_noise_1'
    df[name] = hist[str(col)].map(lambda x: noise_percent(np.array(x)))

print('\nstage 7')

for col in hist:
    print('+', end='')
    name = str(col).split('_')[0] + '_' + str(col).split('_')[1] + '_hist_noise_3'
    df[name] = hist[str(col)].map(lambda x: noise_percent(np.array(x), 3))

print('\nstage 8')

df.filter(regex='face', axis=1).head(5)
```

```python
df_fakes_compact = df[df.fake==True].filter(regex='^(.(?!(raw)))*$', axis=1) \
                                    .filter(regex='^(.(?!(ake)))*$', axis=1) \
                                    .filter(regex='^(.(?!(ilename)))*$', axis=1) \
                                    .filter(regex='^(.(?!(ist_simple)))*$', axis=1)
                                    
df_fakes_reduced = pd.DataFrame(df_fakes_compact.mean()).T
df_fakes_reduced['fake'] = True
df_fakes_reduced = df_fakes_reduced.set_index('fake')
# df_fakes_reduced.filter(like='face', axis=1)
```

```python
df_real_compact = df[df.fake==False].filter(regex='^(.(?!(raw)))*$', axis=1) \
                                    .filter(regex='^(.(?!(ake)))*$', axis=1) \
                                    .filter(regex='^(.(?!(ilename)))*$', axis=1) \
                                    .filter(regex='^(.(?!(ist_simple)))*$', axis=1)
                                    
df_real_reduced = pd.DataFrame(df_real_compact.mean()).T
df_real_reduced['fake'] = False
df_real_reduced = df_real_reduced.set_index('fake')
df_real_reduced
```

```python
df_compare = pd.concat([df_real_reduced, df_fakes_reduced]).T
df_compare['diff_rel'] = abs(df_compare[False] - df_compare[True]) / (df_compare[False] + df_compare[True])
df_compare['diff'] = abs(df_compare[False] - df_compare[True]) 

df_compare = df_compare.sort_values(by=['diff_rel'], ascending=False)
# list(df_compare.head(15).index)
df_compare.head(15)
```

```python
df_fakes_std = pd.DataFrame(df_fakes_compact.std()).T
df_fakes_std['fake'] = True
df_fakes_std = df_fakes_std.set_index('fake')

df_real_std = pd.DataFrame(df_real_compact.std()).T
df_real_std['fake'] = False
df_real_std = df_real_std.set_index('fake')

df_std = pd.concat([df_real_reduced, df_fakes_reduced]).T
df_std['total_std'] = (df_std[False] + df_std[True])

df_std = df_std.sort_values(by=['total_std'], ascending=True)
# df_std.filter(regex='^(.(?!(ist)))*$', axis=0).head(10)
df_std.head(10)
```

```python
# aggregate diff/std

df_smart = pd.concat([pd.DataFrame(df_compare['diff']),pd.DataFrame(df_std['total_std'])], axis=1)
df_smart['metric'] = df_smart['diff'] / df_smart['total_std']
df_smart = df_smart.sort_values(by=['metric'], ascending=False)
# df_smart = df_smart.dropna(subset=['metric'])
df_smart.head(15)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import SGDClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split( 
    df_compact[features], 
    df_compact['fake'],
    test_size=0.33, 
    random_state=63)
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.preprocessing import PolynomialFeatures
```

```python
# ? probabylity = True, predict_proba ?
# clf = make_pipeline(StandardScaler(), SVC(gamma='auto'))
clf = make_pipeline(PolynomialFeatures(degree=2), SVC(gamma='auto'))
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
# y_test = y_train

i = 0
cnt_ok, cnt_fail = 0, 0
cnt_ok_false, cnt_fail_false = 0, 0
for val_test in y_test:
    if y_pred[i] == val_test:
        cnt_ok += 1
        if val_test == False:
            cnt_ok_false += 1
    else:
        if y_pred[i] == True:
            cnt_fail_false += 1
        cnt_fail += 1
    i += 1
        
print(cnt_ok, cnt_fail, cnt_ok_false, cnt_fail_false)
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
target_names = ['class "real"', 'class "fakes"']
print(classification_report(y_test, y_pred, target_names=target_names, zero_division='warn'))
```
