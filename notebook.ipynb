```python
import warnings
warnings.filterwarnings('ignore')

# %%capture --no-stderr --output
print('\ninit')

raw = df.filter(like='_raw', axis=1).copy()

print('\nstage -1')

df['face_size_px'] = raw['overall_face_raw'].map(lambda x: len(x))

print('\nstage 0')

for col in raw:
    print('+', end='')
    name = str(col).split('_')[0] + '_' + str(col).split('_')[1] + '_std'
    df[name] = raw[str(col)].map(lambda x: np.array(x).std())

print('\nstage 1')
    
# raw = df.filter(like='_raw', axis=1).copy()

for col in raw:
    print('+', end='')
    name = str(col).split('_')[0] + '_' + str(col).split('_')[1] + '_var'
    df[name] = raw[str(col)].map(lambda x: np.array(x).var())

print('\nstage 2')

def hist_split(arr):
    res = [0] * 256
    for x in arr:
        res[x] += 1
    return res

# raw = df.filter(like='_raw', axis=1).copy()

for col in raw:
    print('+', end='')
    name = str(col).split('_')[0] + '_' + str(col).split('_')[1] + '_hist_simple'
    df[name] = raw[str(col)].map(lambda x: hist_split(np.array(x)))

print('\nstage 3')

hist= df.filter(like='_hist_simple', axis=1).copy()

for col in hist:
    print('+', end='')
    name = str(col).split('_')[0] + '_' + str(col).split('_')[1] + '_hist_std'
    df[name] = hist[str(col)].map(lambda x: np.array(x).std())

print('\nstage 4')

# hist= df.filter(like='_hist', axis=1).copy()

for col in hist:
    print('+', end='')
    name = str(col).split('_')[0] + '_' + str(col).split('_')[1] + '_hist_var'
    df[name] = hist[str(col)].map(lambda x: np.array(x).var())


print('\nstage 5')

def noise_percent(arr, percent_threshold=1, ignore_zero=True):
    threshold = sum(arr) * percent_threshold / 100
    cnt = 0
    for x in arr:
        if (x < threshold and ((x != 0) or (not ignore_zero))):
            cnt += 1
    return cnt * 100 / float(sum(arr))

for col in hist:
    print('+', end='')
    name = str(col).split('_')[0] + '_' + str(col).split('_')[1] + '_hist_noise_0.5'
    df[name] = hist[str(col)].map(lambda x: noise_percent(np.array(x), 0.5))

print('\nstage 6')

for col in hist:
    print('+', end='')
    name = str(col).split('_')[0] + '_' + str(col).split('_')[1] + '_hist_noise_1'
    df[name] = hist[str(col)].map(lambda x: noise_percent(np.array(x)))

print('\nstage 7')

for col in hist:
    print('+', end='')
    name = str(col).split('_')[0] + '_' + str(col).split('_')[1] + '_hist_noise_3'
    df[name] = hist[str(col)].map(lambda x: noise_percent(np.array(x), 3))

print('\nstage 8 - done')

df.filter(regex='face', axis=1).head(5)
```
